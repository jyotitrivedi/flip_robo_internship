#### 8. Scrape the details of Highest selling novels. Url = https://www.theguardian.com/news/datablog/2012/aug/09/best-selling-books-all-time-fifty-shades-grey-compare/ You have to find the following details: A) Book name B) Author name C) Volumes sold D) Publisher E) Genre

```python
Name=[]
Author=[]
Volume_Sold=[]
Publisher=[]
Genre=[]
for i in range(0,100):
    name=driver.find_element_by_id("table-cell-10943-"+str(i)+"-1")
    Name.append(name.text)
    author=driver.find_element_by_id("table-cell-10943-"+str(i)+"-2")
    Author.append(author.text)
    volume=driver.find_element_by_id("table-cell-10943-"+str(i)+"-3")
    Volume_Sold.append(volume.text)
    publisher=driver.find_element_by_id("table-cell-10943-"+str(i)+"-4")
    Publisher.append(publisher.text)
    genre=driver.find_element_by_id("table-cell-10943-"+str(i)+"-5")
    Genre.append(genre.text)
```

```python
Book=pd.DataFrame()
Book['Name']=Name
Book['Author']=Author
Book['Volume Sold']=Volume_Sold
Book['Publisher']=Publisher
Book['Genre']=Genre
Book
```

#### 10. Details of Datasets from UCI machine learning repositories. Url = https://archive.ics.uci.edu/ You have to find the following details: A) Dataset name B) Data type C) Task D) Attribute type E) No of instances F) No of attribute G) Year

```python
driver=webdriver.Firefox()
```

```python
url="https://archive.ics.uci.edu/"
driver.get(url)
```

```python
all=driver.find_element_by_xpath("//a[contains(text(), 'view all data sets')]").get_attribute('href')
driver.get(all)
```

```python
Dataset=[]
DataType=[]
Task=[]
Attribute=[]
Instances=[]
Nattribute=[]
Year=[]
for i in range(2,590):
    data=driver.find_elements_by_xpath("/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr["+str(i)+"]/td[1]/table/tbody/tr/td[2]/p/b/a")
    for j in data:
        Dataset.append(j.text)
    type=driver.find_elements_by_xpath("/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr["+str(i)+"]/td[2]/p")    
    for j in type:
        DataType.append(j.text)
    task=driver.find_elements_by_xpath("/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr["+str(i)+"]/td[3]/p")    
    for j in task:
        Task.append(j.text)
    attribute=driver.find_elements_by_xpath("/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr["+str(i)+"]/td[4]/p")
    for j in attribute:
        Attribute.append(j.text)
    instances=driver.find_elements_by_xpath("/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr["+str(i)+"]/td[5]/p")
    for j in instances:
        Instances.append(j.text)
    nattribute=driver.find_elements_by_xpath("/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr["+str(i)+"]/td[6]/p")
    for j in nattribute:
        Nattribute.append(j.text)
    year=driver.find_elements_by_xpath("/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr["+str(i)+"]/td[7]/p")
    for j in year:
        Year.append(j.text)
```

```python
UCI=pd.DataFrame()
UCI['DataSet']=Dataset
UCI['Data Types']=DataType
UCI['Task']=Task
UCI['Attribute Types']=Attribute
UCI['# Instances']=Instances
UCI['# Attributes']=Nattribute
UCI['Year']=Year
UCI
```
