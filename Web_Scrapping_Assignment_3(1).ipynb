#### 2. In the above question, now scrape the following details of each product listed in first 3 pages of your search results and save it in a dataframe and csv. In case if any product has less than 3 pages in search results then scrape all the products available under that product vertical. Details to be scraped are: "Brand Name", "Name of the Product", "Rating", "No. of Ratings", "Price", "Return/Exchange", "Expected Delivery", "Availability", "Other Details" and “Product URL”. In case, if any of the details are missing for any of the product then replace it by “-“.

```python
urls=[]
prev=driver.find_element_by_xpath("//ul[@class='a-pagination']/li[2]/a").get_attribute('href')
urls.append(prev)
for i in range(0,2):
    next=driver.find_element_by_xpath("//li[@class='a-last']/a").get_attribute("href")
    urls.append(next)
    driver.get(next)
```

```python
urls
```

```python
product_url=[]
for i in urls:
    driver.get(i)
    time.sleep(3)
    UR=driver.find_elements_by_xpath("//a[@class='a-link-normal s-no-outline']")
    for j in UR:
        product_url.append(j.get_attribute("href"))
```

```python
len(product_url)
```

```python
from selenium.common.exceptions import NoSuchElementException,TimeoutException
Brand_Name=[]
Product=[]
Rating=[]
No_of_Ratings=[]
Price=[]
Return_Exchange=[]
Expected_Delivery=[]
Availability=[]
Other_Details=[]
for i in product_url:
    try:
        driver.get(i)
        try:
            brand=driver.find_element_by_id('bylineInfo')
            Brand_Name.append(brand.text)
        except NoSuchElementException:
            Brand_Name.append('-') 
        try:    
            title=driver.find_element_by_id('productTitle')
            Product.append(title.text)
        except NoSuchElementException:
            Product.append('-') 
        try:    
            rating=driver.find_element_by_xpath("//span[@class='a-size-medium a-color-base']")
            Rating.append(rating.text)
        except NoSuchElementException:
            Rating.append('-')     
        try:
            price=driver.find_element_by_id('priceblock_ourprice')
            Price.append(price.text)
        except NoSuchElementException:
            Price.append('-') 
        try:
            exp=driver.find_element_by_xpath("//div[@class='a-section a-spacing-mini']/b")
            Expected_Delivery.append(exp.text)
        except NoSuchElementException:
            Expected_Delivery.append('-') 
        try:
            ret=driver.find_element_by_xpath("//div[@class='a-section a-spacing-none icon-content']/a")
            Return_Exchange.append(ret.text)
        except NoSuchElementException:
            Return_Exchange.append('-') 
        try:    
            avail=driver.find_element_by_xpath("//div[@id='availability']/span")
            Availability.append(avail.text)
        except NoSuchElementException:
            Availability.append('-') 
        try:
            details=driver.find_element_by_xpath("//ul[@class='a-unordered-list a-vertical a-spacing-mini']")
            Other_Details.append(details.text)
        except NoSuchElementException:
            Other_Details.append('-') 
        try:
            nor=driver.find_element_by_id('acrCustomerReviewText')
            No_of_Ratings.append(nor.text)
        except NoSuchElementException:
            No_of_Ratings.append('Not Available')
    except TimeoutException:
        continue
    
    
    
```

```python
Amazon=pd.DataFrame()
Amazon['Brand']=Brand_Name
Amazon['Product']=Product
Amazon['Rating']=Rating
Amazon['Price']=Price
Amazon['No. Of Ratings']=No_of_Ratings
Amazon['Expected Delivery']=Expected_Delivery
Amazon['Return/Exchange']=Return_Exchange
Amazon['Availability']=Availability
Amazon['Details']=Other_Details
Amazon[50:150]
```

#### 9. Write a program to extract at least 500 Comments, Comment upvote and time when comment was posted from any YouTube Video.

```python
driver=webdriver.Firefox()
url='https://www.youtube.com/watch?v=Xn7KWR9EOGQ'
driver.get(url)
```

```python
Time=[]
Upvote=[]
Comments=[]
while(True):
    html = driver.find_element_by_tag_name('html')
    html.send_keys(Keys.PAGE_DOWN)
    comment=driver.find_elements_by_xpath("//div[@id='content']/yt-formatted-string[2]")
    for i in comment:
        Comments.append(i.text)
    upvote=driver.find_elements_by_id('vote-count-middle')
    for i in upvote:
        Upvote.append(i.text)
    tim=driver.find_elements_by_xpath("//div[@id='header-author']/yt-formatted-string/a")
    for i in tim:
        Time.append(i.text)
    if len(Comments)>500:
        break
    
```

```python
driver=webdriver.Firefox()
url='https://www.hostelworld.com/s?q=London,%20England&country=England&city=London&type=city&id=3&from=2021-07-06&to=2021-07-09&guests=2&page=1'
driver.get(url)
```

```python

Distance=[]

Private_Price=[]
Dorm_Price=[]

dist=driver.find_elements_by_xpath("//div[@class='subtitle body-3']/a/span[1]")
for i in dist:
    Distance.append(i.text)
try:    
    pp=driver.find_elements_by_xpath("//a[@class='prices']/div[1]/div") 
    for i in pp:
        Private_Price.append(i.text)
except NoSuchElementException:
    Private_Price.append('-')
try:
    dp=driver.find_elements_by_xpath("//a[@class='prices']/div[2]/div") 
    for i in dp:
        Dorm_Price.append(i.text)
except NoSuchElementException:
    Dorm_Price.append('-')
 
```

```python
urls=[]
url=driver.find_elements_by_xpath("//h2[@class='title title-6']/a")
for i in url:
    urls.append(i.get_attribute('href'))   
```

```python
Name=[]
Ratings=[]
Total_Review=[]
Overall_Review=[]
Facilities=[]
Desc=[]
for i in urls[:]:
    driver.get(i)
    desc=driver.find_elements_by_xpath("//div[@class='content collapse-content']")
    for i in desc:
        Desc.append(i.text)
    fac=driver.find_elements_by_xpath("//ul[@class='facility-sections']")
    for i in fac:
        Facilities.append(i.text)
    try:
        rate=driver.find_elements_by_xpath("//div[@class='score orange big']")
        for i in rate:
            Ratings.append(i.text)
    except NoSuchElementException:
        Ratings.append('-')
    name=driver.find_elements_by_xpath("//div[@class='title-2']")
    for i in name:
        Name.append(i.text)
    rev=driver.find_elements_by_xpath("//div[@class='reviews']")
    for i in rev:
        Total_Review.append(i.text)
    over=driver.find_elements_by_xpath("//div[@class='keyword']/span")
    for i in over:
        Overall_Review.append(i.text)
    try:    
        pp=driver.find_elements_by_xpath("//a[@class='prices']/div[1]/div") 
        for i in pp:
            Private_Price.append(i.text)
    except NoSuchElementException:
        Private_Price.append('-')
    try:
        dp=driver.find_elements_by_xpath("//a[@class='prices']/div[2]/div") 
        for i in dp:
            Dorm_Price.append(i.text)
    except NoSuchElementException:
        Dorm_Price.append('-')
     
       
```
